```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/pokemon-dataset/Pokemon.csv")
```

```python
df.head()
```

```python
df.describe()
```

# Bar plot of Pokemon Types: Show the distribution of Pokemon types

```python
# Countplot for Pokemon types
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='type1', order=df['type1'].value_counts().index, palette='colorblind')
plt.title('Distribution of Pokemon Types (Type 1)')
plt.xlabel('Type 1')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Scatter plot of Attack vs. Defense: Explore the relationship between attack and defense stats

```python
# Scatter plot of Attack vs. Defense
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='attack', y='defense', hue='legendary', palette='coolwarm', alpha=0.8)
plt.title('Attack vs. Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.legend(title='Legendary', loc='upper right')
plt.show()
```

# Box plot of Total stats by Generation: See how the total stats vary across different generations

```python
# Box plot of Total stats by Generation
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='generation', y='total', palette='Set2')
plt.title('Total Stats by Generation')
plt.xlabel('Generation')
plt.ylabel('Total Stats')
plt.show()
```

# Distribution of Legendary Pokemon: Visualize the proportion of legendary vs. non-legendary Pokemon

```python
# Pie chart of Legendary vs. Non-legendary Pokemon
legendary_counts = df['legendary'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(legendary_counts, labels=legendary_counts.index, autopct='%1.2f%%', colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Legendary Pokemon')
plt.show()
```

# Histogram of HP distribution: Understand the distribution of Hit Points (HP) among Pokemon

```python
# Histogram of HP distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='hp', bins=20, kde=True, color='orange')
plt.title('Distribution of Hit Points (HP)')
plt.xlabel('Hit Points (HP)')
plt.ylabel('Frequency')
plt.show()
```

# Pairplot of Stats: Explore pairwise relationships between different stats

```python
# Pairplot of stats
sns.pairplot(df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']])
plt.show()
```

# Bar plot of Average Total Stats by Type: Show the average total stats for each type

```python
# Bar plot of Average Total Stats by Type
avg_total_by_type = df.groupby('type1')['total'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_total_by_type.index, y=avg_total_by_type.values, palette='viridis')
plt.title('Average Total Stats by Type (Type 1)')
plt.xlabel('Type 1')
plt.ylabel('Average Total Stats')
plt.xticks(rotation=45)
plt.show()
```

# Heatmap of Correlation Matrix: Visualize the correlation between different stats

```python
# Heatmap of Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Pokemon Stats')
plt.show()
```

# Box plot of Attack by Type: See how attack stats vary across different types

```python
# Box plot of Attack by Type
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='type1', y='attack', palette='Pastel1')
plt.title('Attack by Type (Type 1)')
plt.xlabel('Type 1')
plt.ylabel('Attack')
plt.xticks(rotation=45)
plt.show()
```

# Violin plot of Defense by Generation: Visualize the distribution of defense stats across different generations

```python
# Violin plot of Defense by Generation
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='generation', y='defense', palette='Set3')
plt.title('Defense by Generation')
plt.xlabel('Generation')
plt.ylabel('Defense')
plt.show()
```
